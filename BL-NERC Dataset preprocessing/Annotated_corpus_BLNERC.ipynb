# Bongkim

```python
print('total sentence bongkim:',df1['Sentence #'].nunique(),'\n')
```

```python
bon_word_list=[]
bon_tag_list=[]
bon_sent_list=[]
```

```python
class SentenceGetter(object):
    
    def __init__(self, data):
        self.n_sent = 1
        self.data = data
        self.empty = False
        agg_func = lambda s: [(w, t) for w, t in zip(s["Word"].values.tolist(),
                                                           s["Tag"].values.tolist())]
        self.grouped = self.data.groupby("Sentence #").apply(agg_func)
        self.sentences = [s for s in self.grouped]
    
    def get_next(self):
        try:
            s = self.grouped["Sentence: {}".format(self.n_sent)]
            self.n_sent += 1
            return s
        except:
            return None
```

```python
getter = SentenceGetter(df1)
```

```python
bon_sent = getter.sentences
bon_sent[0]
```

```python
for i in range(len(bon_sent)):
    bon_sent_list.append(i)
```

```python
bon_data = pd.DataFrame()

bon_data['Sentence_no']= bon_sent_list
```

```python
bon_data
```

```python
for i in range(len(bon_data)):
    objective_ouput_list = [item[0] for item in bon_sent[i]]
    bon_word_list.append(objective_ouput_list)
```

```python
bon_data['words'] = bon_word_list 
bon_data
```

```python
for i in range(len(bon_data)):
    objective_ouput_list = [item[1] for item in bon_sent[i]]
    bon_tag_list.append(objective_ouput_list)
```

```python
bon_data['tags'] = bon_tag_list
bon_data
```

```python
bon_tr = bon_data.iloc[:1600].reset_index(drop=True)
bon_ts = bon_data.iloc[1600:1800].reset_index(drop=True)
bon_val = bon_data.iloc[1800:].reset_index(drop=True)
print(' bongkim train:',bon_tr.shape,'\n','bongkim test:',bon_ts.shape,'\n','bongkim valid:',bon_val.shape)
```

```python
shotto_word_list=[]
shotto_tag_list=[]
shotto_sent_list=[]

class SentenceGetter(object):
    
    def __init__(self, data):
        self.n_sent = 1
        self.data = data
        self.empty = False
        agg_func = lambda s: [(w, t) for w, t in zip(s["Word"].values.tolist(),
                                                           s["Tag"].values.tolist())]
        self.grouped = self.data.groupby("Sentence #").apply(agg_func)
        self.sentences = [s for s in self.grouped]
    
    def get_next(self):
        try:
            s = self.grouped["Sentence: {}".format(self.n_sent)]
            self.n_sent += 1
            return s
        except:
            return None

getter = SentenceGetter(df12)
shotto_sent = getter.sentences
shotto_sent[0]
```

```python
for i in range(len(shotto_sent)):
    shotto_sent_list.append(i)

shotto_data = pd.DataFrame()
shotto_data['Sentence_no']= shotto_sent_list

for i in range(len(shotto_data)):
    objective_ouput_list = [item[0] for item in shotto_sent[i]]
    shotto_word_list.append(objective_ouput_list)

shotto_data['words'] = shotto_word_list 

for i in range(len(shotto_data)):
    objective_ouput_list = [item[1] for item in shotto_sent[i]]
    shotto_tag_list.append(objective_ouput_list)

shotto_data['tags'] = shotto_tag_list
shotto_data
```

```python
shotto_tr = shotto_data.iloc[:1600].reset_index(drop=True)
shotto_ts = shotto_data.iloc[1600:1800].reset_index(drop=True)
shotto_val = shotto_data.iloc[1800:].reset_index(drop=True)
print(' shottojit train:',shotto_tr.shape,'\n','shottojit test:',shotto_ts.shape,'\n','shottojit valid:',shotto_val.shape)
```

# To JSON

```python
dataset = { 
    'train' : [],
    'test' : [],
    'validation' : []
    }
```

```python
def conv_to_dict(df, df_name):
    
    for i in range(len(df)):
        id = i
        ner_tag = df['ner_tags'][i]
        token =  df['words'][i]
    
        temp = {
            
            'id' : str(id),
            'ner_tags': ner_tag,
            'tokens': token
            
        }
        

        dataset[df_name].append(temp)
        
```

```python
conv_to_dict(data_tr, 'train')
```

```python
conv_to_dict(data_ts, 'test')
```

```python
conv_to_dict(data_val, 'validation')
```

```python
import json

s = json.dumps(dataset)

with open('BLNERC dataset new.json','w') as f:
    f.write(s)
```
