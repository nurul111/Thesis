```python
import pandas as pd
import nltk
```

```python
df=pd.read_excel(r"D:\Thesis\Ner\Dataset\author_random_sentence\zahir_rayhan_random.xlsx")
df
```

```python
df1=df['sent']
```

```python
from nltk.tokenize import word_tokenize
from nltk.tokenize import sent_tokenize
token=[]
sent=[]
tag=[]
sentcount=0
for i in df1:
    words=word_tokenize(i)
    sentcount=sentcount+1
    for j in range(len(words)):
        token.append(words[j])
        tag.append('O')
        sent.append(f'sentence:{sentcount}')
        #else:sent.append('NaN')
df2=pd.DataFrame(token)
df2.columns=['word']
df3 = pd.DataFrame(sent)
df3.columns=['Sentence #']
df4 = pd.DataFrame(tag)
df4.columns=['Tag']
df3
    
```

```python
df=pd.concat([df3,df2,df4],axis=1)
df
```
